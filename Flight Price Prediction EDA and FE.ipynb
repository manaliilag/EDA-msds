```python
def duration_to_min(x):
    if ' ' in x:        
        a = x.split(" ")
        a_hour = x.split(" ")[0].split("h")[0]
        a_min = x.split(" ")[1].split("m")[0]
        total_mins = int(a_hour) * 60 + int(a_min)        
    elif 'h' in x:
        a = x.split("h")
        a_hour = x.split("h")[0]
        total_mins = int(a_hour) * 60        
    elif 'm' in x:
        a = x.split("m")
        total_mins = x.split("m")[0]
    return total_mins            
```
