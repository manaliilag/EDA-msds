### Problem Statement
A retail compny "ABC Private Limited" wants to understand the customer purchase behaviour (specilly purchase amount) against various products of different categories, they have shared purchase summary of various customers for selected high volume products from last month the data set also contains customer demographics (age, gender, marital status, city typr, stay in current city) product details (product_id and product category) and total purchase_amount from last month 

Now they want to build a model to predict the purchase amount of customer against various  products which will help them to create personlized offer for customers against different products
